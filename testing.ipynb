```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error
```

```python
df=pd.read_csv('expense_data.csv')
```

```python
df.head(5)
```

```python
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
df['Account']=pd.to_numeric(df['Account'],errors='coerce')
```

```python
df=df[df['Income/Expense'].str.lower()=='expense']
```

```python
daily=df.groupby(df['Date'].dt.date)['Amount'].sum()
```

```python
dates=pd.date_range(min(daily.index),max(daily.index))
```

```python
daily=daily.reindex(dates, fill_value=0)
```

```python
data=pd.DataFrame({'expense':daily})
```

```python
data['lag1']=data['expense'].shift(1)
data['lag2']=data['expense'].shift(2)
data['lag3']=data['expense'].shift(3)
data['lag7']=data['expense'].shift(7)
```
